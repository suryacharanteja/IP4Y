```python
# binary
x = 0
while x <= 1000:
    binary_number = bin(x)  # Convert x to binary
    decimal_number = int(binary_number, 2)  # Convert octal_number to decimal
    print(f"Octal: {binary_number}, Decimal: {decimal_number}")
    x += 1
```

```python
octal
x = 0
while x <= 1000:
    octal_number = oct(x)  # Convert x to octal
    decimal_number = int(octal_number, 8)  # Convert octal_number to decimal
    print(f"Octal: {octal_number}, Decimal: {decimal_number}")
    x += 1
```

```python
# hexadecimal
x = 0
while x <= 10000:
    hex_number = hex(x)  # Convert x to hexadecimal
    decimal_number = int(hex_number, 16)  # Convert hex_number to decimal
    print(f"Octal: {hex_number}, Decimal: {decimal_number}")
    x += 1
```
